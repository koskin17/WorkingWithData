# Фильтры

```python
data = pd.read_csv('power.csv')
data.head()
```

```python
# выбрать несколько столбцов

country_stats = data.filter(items = ['country', 'quantity'])
country_stats.head()
```

```python
# или так

data[['country', 'quantity']].head()
```

### Отфильтруем строки с потреблением выше среднего

```python
average_level = data['quantity'].mean()
average_level
```

```python
'quantity > {}'.format(average_level)
```

```python
# Первый способ фильтрации строк для получения только тех, которые выше среднего
average_level = data['quantity'].mean()
country_stats.query('quantity > {}'.format(average_level)).head()
```

```python
# Второй способ фильтрации строк, когда в квадратных скобках [] указывается условие.
data[ data.quantity > average_level ].head()
```

### Как определить используемый вариант названия страны?

```python
data['country'].unique()
```

```python
# найдем как называется Россия и Беларусь в этом датафрейме
# фильтр на подстроку - смотрим все страны, содержащие в названии 'us'

data[ data['country'].str.contains('kra', case=False) ]['country'].unique()
```

```python
# фильтр на несколько условий сразу
# | - условие ИЛИ
# & AND
# () | (() | () & ())
# Перечислением условий фильтров удобно фильтровать сразу несколько столбцов.
filtered_countries = data[ (data['country']=='Ukraine') | (data['country']=='Poland') ]

filtered_countries
```

```python
# Также строки можно фильтровать методом isin()
# Этот метод удобен, если надо фильтровать один столбец по нескольких значениям.
data[data.country.isin(['Ukraine', 'Poland'])]
```

```python
filtered_countries['country'].unique()
```

```python
# фильтры на номер строки. Важно помнить, что под номером строки понимается индекс.
# Также важно помниить, что метод loc в pandas правую границу берёт, т.е. строк будет не 5, а 6.

data.loc[1000:1005]
```

```python
# сортировка по убыванию

data.sort_values('quantity', ascending=False).head()
```

```python
# сортировка по нескольким столбцам. При этом для каждого столбца можно указывать принцип сортировки: по убыванию или по возрастанию через указание параметра ascending = True или False. Если столбцов несколько, то в порядке их перечисления указывается и параметр True / False, которые будут применены к соответствующим столбцам по порядку.

data.sort_values(by=['year', 'country', 'quantity'], ascending=[False, True, False]).head(50)
```

```python
data.head()
```

### Упражнение
Используем transactions.csv.

Для какой транзакции (столбец ID) были наибольшие расходы (столбец Cost) в категории "_8" (столбец Product)? 

```python
# dataframe.sort_values('Cost', ascending=False).filter('Product'='_8')
filtered_product = dataframe[(dataframe['Product'] == '_8')]
filtered_product.sort_values('Cost', ascending=False)
```

```python
# Вариант с .loc
filtered_product = (dataframe
                    .loc[dataframe['Product'] == '_8']
                    .sort_values('Cost', ascending=False))
filtered_product
```

```python
# Вариант с query
filtered_product = dataframe.query("Product == '_8'").sort_values('Cost', ascending=False)
filtered_product
```

```python
# Вариант с ограничением строк
top10 = dataframe.loc[dataframe['Product'] == '_8'].nlargest(1, 'Cost')
top10
```

```python
# Вариант с выбором колонок для вывода
filtered_product = (dataframe.loc[dataframe['Product'] == '_8', ['Product', 'Cost']]
                    .sort_values('Cost', ascending=False))
filtered_product
```

```python
# Если столбец Product часто фильтруется и содержит много уникальных значений, то рекомендуется его преобразовать в категорию, что позволит ускорить фильтрацию и уменьшить потребление памяти.
dataframe['Product'] = dataframe['Product'].astype('category')
filtered_product = (dataframe.loc[dataframe['Product'] == '_8', ['Product', 'Cost']]
                    .sort_values('Cost', ascending=False))
filtered_product
```
